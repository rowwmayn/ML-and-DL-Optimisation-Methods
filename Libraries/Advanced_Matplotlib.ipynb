```python
# Matplotlib is a data visualisation library.
import numpy as np;
import matplotlib.pyplot as plt;
print("Zebra");
```

```python
x_data = np.random.random(50)*100;
y_data = np.random.random(50)*100;
```

Scatterplot

```python
plt.scatter(x_data,y_data, c="blue", marker="*", s=150, alpha=0.3);
plt.show();
```

```python
# Time series

year = [2006 + x for x in range(16)];
weight = [80 + x for x in range(16)];

#print(year);
#print(weight);

plt.plot(year,weight, c='r', lw='5', linestyle="--");
```

```python
# Bar charts
x = ["C++", "C", "Java", "Python", "Zebra"];
y = [20, 11, 28, 57, 12];
plt.xlabel("Languages");
plt.ylabel("Users", color="r");
plt.title("Users of different languages");
plt.bar(x,y, color="r", align='edge', width=0.5, edgecolor="green", lw=0.6); 
```

```python
# Histograms (Shows frequencies)
ages = np.random.normal(20, 11.5, 10000000);
plt.hist(ages);
plt.show();
```
