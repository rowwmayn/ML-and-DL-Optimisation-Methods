```python
labels = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE',
    'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'
];
df = pd.read_csv(r"F:\Personal Projects\ML-and-DL-Optimisation-Methods\Datasets\housing.csv", sep=r"\s+", header=None, encoding="utf-8", engine='python');
df_clone = df.copy();
df_clone.columns = labels;
#print(df_clone.head(10));



df = df.to_numpy();
ones = np.ones((df.shape[0],1));
df_designed=np.hstack((ones, df));
row,col = df_designed.shape;
#print(epoch,col);
#print(df_designed.shape);
X = df_designed[:,0:df_designed.shape[1]-1];
#print(X[0][-1])
Y = df_designed[:,df_designed.shape[1]-1].reshape(row,1);
#print(Y.shape);
#print(Y);



# Initialisations!
grad = np.zeros(Y.shape); 
#li = [i for i in range(506)];
random_int = np.random.randint(0,row);
#print(random_int);
theta = [];
step_size = 0.00001;
errors = [];



train,test = train_test_split(df_designed,test_size=0.2);

```

```python
# 1. Initialize theta before loops
theta = np.zeros((X.shape[1], 1)); 
step_size = 0.000001; 

for e in range(100): # Epoch loop
    # 2. Use the SHUFFLED data for this pass
    shuffled_df = np.random.permutation(train);
    
    epoch_loss = 0;
    step_size = step_size / (1 + 0.1 * i)
    for i in range(shuffled_df.shape[0]): # Row loop
        curr = shuffled_df[i];
        
        # Features (everything but last column)
        xi = curr[:-1].reshape(1, -1); # Shape (1, 14)
        yi = curr[-1];
        
        # 3. Predict & Error
        pred_y = xi @ theta;
        error = pred_y - yi;
        
        # 4. Square the loss for monitoring
        epoch_loss += error**2;
        
        # 5. Calculate Gradient for ONE sample
        # grad = 2 * error * xi_transpose
        grad = 2 * error * xi.T; 
        
        # 6. THE CORE OF SGD: Update theta IMMEDIATELY
        theta = theta - step_size * grad;
        
    # Print average loss for this epoch
    print(f"Epoch {e}, MSE: {epoch_loss/shuffled_df.shape[0]}");
```
