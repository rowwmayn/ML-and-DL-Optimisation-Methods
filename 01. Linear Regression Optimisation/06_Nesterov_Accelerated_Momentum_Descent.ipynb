```python
import numpy as np;
import matplotlib.pyplot as plt;
from sklearn.model_selection import train_test_split as tts;
import pandas as pd;
```

```python
column_names = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE',
    'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'
];
full_data = pd.read_csv(
    r"F:\Personal Projects\ML-and-DL-Optimisation-Methods\Datasets\housing.csv",
    sep=r'\s+', header=None, names=column_names, encoding="utf-8", engine='python'
);

#data_matrix = full_data.to_numpy;
```

```python
def split_data_add_bias(df):
    data_matrix = df.to_numpy();
    r,c = data_matrix.shape;
    ones = np.ones((r,1));

    data_matrix = np.hstack((ones, data_matrix));

    X = data_matrix[:,0:c-2];
    Y = data_matrix[:,-1].reshape(r,1);
    X_train, X_test, Y_train, Y_test = tts(X, Y, train_size=0.8, random_state=5809);
    return X_train, X_test, Y_train, Y_test;
```

```python
def Nesterov_Accelerated_Momentum(df):
    X_train, X_test, Y_train, Y_test = split_data_add_bias(df);
    # Initialisations
    beta = 0.91;
    alpha = 0.00001;
    iterations = 50000;
    m,n = X_train.shape;
    theta = np.zeros((n,1));
    v = np.zeros((n,1));

    # Storing for visualisations
    thetas = [];
    la_thetas = [];
    n_vs = [];
    grads = [];


    # Algorithm
    for i in range(iterations):
        theta_la = theta - (beta*alpha*v);
        #la_thetas.append(theta_la);

        grad = (1/m)*(X_train.T@(X_train@theta_la - Y_train)); # Gradient using lookahead theta
        #grads.append(grad);
        v = beta*v + (1-beta)*grad;
        #n_vs.append(v);

        theta = theta - alpha*v;
        #thetas.append(theta);

        thetas.append(theta.copy());
        la_thetas.append(theta_la.copy());
        grads.append(grad.copy());
        n_vs.append(v.copy());

        if i%200==0:
            print(f"Iteration {i} and current theta: {theta}");

    return thetas, la_thetas, n_vs, grads, X_train, X_test, Y_train, Y_test;
```

```python
thetas, la_thetas, n_vs, grads, X_train, X_test, Y_train, Y_test = Nesterov_Accelerated_Momentum(full_data);
```

```python
def R2(thet, X_test, Y_test):

    Y_test = Y_test.reshape(-1,1);

    pred = X_test @ thet;
    e = Y_test - pred;

    rss = np.sum(e**2);

    y_mean = np.mean(Y_test);
    tss = np.sum((Y_test - y_mean)**2);

    r2 = 1 - (rss / tss);

    return r2;

```
